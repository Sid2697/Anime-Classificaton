```python
from keras.models import load_model, Model
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
model = load_model('/Users/siddhantbansal/Desktop/Project_files/Siddhant_Code/models/5_layers_final/second_model_final.h5')
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
# Imporing necessary packages
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [30, 30]
import numpy as np
import glob
import cv2
```

```python
path = '/Users/siddhantbansal/Desktop/Project_files/fe46b0f1f7cb6f453df3339abcf76472.jpg'
image = cv2.imread(path)
image = cv2.resize(image, (128, 128))
image = image.reshape(1, 128, 128, 3)
print('Shape of the image is:', image.shape)
plt.imshow(image.reshape(128, 128, 3))
```

```python
clip14 = Model(model.inputs, model.layers[-8].output)
p14 = clip14.predict(image)
print(p14.shape)
p14 = p14.reshape(30, 14, 14)
for i in range(1, 31):
    plt.subplot(6, 5, i)
    plt.imshow(p14[i-1])
```
